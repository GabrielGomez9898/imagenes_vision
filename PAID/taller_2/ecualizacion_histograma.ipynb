```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image_quito = cv2.imread("images/QUITO.png", cv2.IMREAD_GRAYSCALE)

# Visualizar imagen
plt.imshow(image_quito, cmap="gray",  vmin=0, vmax=255)
plt.show()

# Visualizar histograma
hist = cv2.calcHist([image_quito],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
# Ecualizar el histograma
imagen_equ = cv2.equalizeHist(image_quito)

# Visualizar imagen
plt.imshow(imagen_equ, cmap="gray",  vmin=0, vmax=255)
plt.show()

# Visualizar histograma
hist = cv2.calcHist([imagen_equ],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

¿a qué se debe la discontinuidad del histograma ecualizado?
Es discontinuo porque se reasignaron los valores de los pixeles desde 0 hasta 255. Como cada valor se reclasificó con el objetivo de que todos tuvieran la misma probabilidad, entonces quedaron espacios vacios que es donde el histograma es discontinuo. 


```python
# Efectúe una segunda vez esta operación
# Ecualizar el histograma 2
imagen_equ2 = cv2.equalizeHist(imagen_equ)

# Visualizar imagen
plt.imshow(imagen_equ2, cmap="gray",  vmin=0, vmax=255)
plt.show()

# Visualizar histograma
hist = cv2.calcHist([imagen_equ2],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

Efectúe una segunda vez esta operación. Es decir, aplique una ecualización del histograma a la imagen resultado del punto anterior. ¿Qué pasa? ¿Por qué?

Aunque son muy parecidos los dos histogramas, tienen unas ligeras diferencias, lo que sucede es que en la ecualizacion se reasignan los valores de los pixeles discretamente, es decir que se hacen muchas aproximaciones, es por eso que quedan espacios en blanco. Estas aproximaciones vuelven la ecualizacion una funcion idempotente, si se vuelve a aplicar algunos grises pueden reclasificarse neuvamente.

```python
# Repita las operaciones 1 a 4 con la imagen baboon.png
image_baboon = cv2.imread("images/baboon.png", cv2.IMREAD_GRAYSCALE)

# Visualizar imagen
plt.imshow(image_baboon, cmap="gray",  vmin=0, vmax=255)
plt.show()

# Visualizar histograma
hist = cv2.calcHist([image_baboon],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

# Ecualizar el histograma
imagen_equ_baboon= cv2.equalizeHist(image_baboon)

# Visualizar imagen
plt.imshow(imagen_equ_baboon, cmap="gray",  vmin=0, vmax=255)
plt.show()

# Visualizar histograma
hist = cv2.calcHist([imagen_equ_baboon],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

# Efectúe una segunda vez esta operación
# Ecualizar el histograma 2
imagen_equ_baboon_2 = cv2.equalizeHist(imagen_equ_baboon)

# Visualizar imagen
plt.imshow(imagen_equ_baboon_2, cmap="gray",  vmin=0, vmax=255)
plt.show()

# Visualizar histograma
hist = cv2.calcHist([imagen_equ_baboon_2],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

¿Cómo es el resultado de esta transformación comparado con la ecualización de la imagen precedente (quito.png)? ¿Por qué?

La diferencia es que el histrograma ecualizado del baboon tiene menos espacios, esto se debe a que el baboon tiene mas grises y estan mas distribuidos en su imagen original (mas expandido), entonces al reasignar los valores, van a ocupar mas niveles de grises por lo tanto no habrán tantos espacios en blanco.
