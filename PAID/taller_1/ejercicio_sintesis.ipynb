```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_muscle = cv2.imread("images/MUSCLE.png", cv2.IMREAD_GRAYSCALE)

# Visualizar imagen (MUSCLE)
plt.imshow(image_muscle, cmap="gray")
plt.show()

# Visualizar histograma 
hist = cv2.calcHist([image_muscle],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
# Cambiar las fibras rojas (grises mas oscuros) a blanco

M , N = image_muscle.shape

new_muscle = image_muscle.copy()

for i in range(M):
    for j in range(N):
        if image_muscle[i,j] < 120:
            new_muscle[i,j] = 255
        else:
            new_muscle[i,j] = image_muscle[i,j]
            
# Visualizar imagen (MUSCLE)
plt.imshow(new_muscle, cmap="gray")
plt.show()


# ------ Otra implementacion -----
# Crear máscara de fibras rojas (grises oscuros)
mask = (image_muscle < 120).astype('uint8') * 255
plt.imshow(mask, cmap="gray")
plt.title("mascara")
plt.show()

new_muscle_3 = cv2.bitwise_or(image_muscle,mask,0)
plt.imshow(new_muscle_3, cmap="gray")
plt.title("imagen 3")
plt.show()

            
```
