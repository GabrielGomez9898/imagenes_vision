```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_regle = cv2.imread("images/REGLE.png", cv2.IMREAD_GRAYSCALE)

# Visualizar imagen (REGLE)
plt.imshow(image_regle, cmap="gray")
plt.show()

# Visualizar histograma (image_regle)
hist = cv2.calcHist([image_regle],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
# Calcular negativo
M = image_regle.shape[0]
N = image_regle.shape[1]

print(image_regle.shape)
print(M)
print(N)

image_neg = cv2.bitwise_not(image_regle)

# Negativo de forma manual (esto es para nosotros para estudiar) 
#for i in range(M):
#    for j in range(N):
#        image_regle[i, j] = 255 - image_regle[i, j]
            
# Visualizar imagen (REGLE)
plt.imshow(image_neg, cmap="gray")
plt.show()

# Visualizar histograma (image_neg)
hist = cv2.calcHist([image_neg],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```
