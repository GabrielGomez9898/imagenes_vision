```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread("images/amour.png", cv2.IMREAD_GRAYSCALE)

# Visualizar imagen
plt.imshow(image, cmap="gray",  vmin=0, vmax=255)
plt.show()

# Visualizar histograma
hist = cv2.calcHist([image],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

```

```python
from sklearn.cluster import KMeans
import numpy as np

# Segmentacion por agrupacion: k-means K=2
N, M = image.shape

img_flat = image.reshape(N*M, 1)

model_kmeans = KMeans(n_clusters=2)
model_kmeans.fit(img_flat)

labels = model_kmeans.labels_

segmented_image = labels.reshape(N, M)

plt.imshow(segmented_image, cmap="gray")
plt.show()
```

```python
# Segmentacion por umbralizacion
# Aplique el algoritmo de umbralización global
thres_value, thres_image = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

# Visualizar imagen
plt.imshow(thres_image, cmap="gray",  vmin=0, vmax=255)
plt.show()

# Visualizar histograma
hist = cv2.calcHist([thres_image],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```
