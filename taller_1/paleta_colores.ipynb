```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Genere  una imagen de 256x256x1 con valores entre 255 y 0
image = np.zeros((256, 256), dtype=np.uint8)

M, N = image.shape

increment = 1

for i in range(M):
    extra = 0
    for j in range(N):
        image[i,j] = image[i,j]+extra
        extra = extra + increment

# Visualizar
plt.imshow(image, cmap="gray", vmin=0, vmax=255)
plt.show()
```

```python
# Cargar imagenes de trabajo

image_aquitain = cv2.imread("images/AQUITAIN.png", cv2.IMREAD_GRAYSCALE)
image_scanner = cv2.imread("images/SCANNER.png", cv2.IMREAD_GRAYSCALE)

# Visualizar imagen (image_aquitain)
plt.imshow(image_aquitain)
plt.show()

# Visualizar imagen (image_scanner)
plt.imshow(image_scanner)
plt.show()
```

```python
# Aplicar mapa de color a scanner
lut_inverse = np.zeros((256, 1), dtype=np.uint8)
for i in range(0,256):
    lut_inverse[i] = 255 - i


dstImage = cv2.LUT(image_scanner , lut_inverse )
# Visualizar imagen
plt.imshow(dstImage)
plt.show()

lut_color = np.zeros((256, 1, 3), dtype=np.uint8)
for i in range(256):
    lut_color[i, 0, 0] = i  
    lut_color[i, 0, 1] = 0   
    lut_color[i, 0, 2] = 0   

dstImage = cv2.LUT(cv2.merge([dstImage , dstImage , dstImage ]), lut_color)
# Visualizar imagen
plt.imshow(dstImage)
plt.show()
```

```python
# En la imagen scanner.png realice una expansión lineal del contraste

# Expansion del contraste (normalizar) del histograma
M, N = image_scanner.shape

image_exp = cv2.normalize(
    image_scanner,
    None,
    0,
    255,
    cv2.NORM_MINMAX
)

# Visualizar imagen
plt.imshow(image_exp)
plt.show()

# Visualizar histograma (image_exp)

hist = cv2.calcHist([image_exp],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

# ---- Aplicar mapa de color a scanner expandido ------
lut_inverse = np.zeros((256, 1), dtype=np.uint8)
for i in range(0,256):
    lut_inverse[i] = 255 - i


dstImage = cv2.LUT(image_exp , lut_inverse )
# Visualizar imagen
plt.imshow(dstImage)
plt.show()

lut_color = np.zeros((256, 1, 3), dtype=np.uint8)
for i in range(256):
    lut_color[i, 0, 0] = i  
    lut_color[i, 0, 1] = 0   
    lut_color[i, 0, 2] = 0   

dstImage = cv2.LUT(cv2.merge([dstImage , dstImage , dstImage ]), lut_color)
# Visualizar imagen
plt.imshow(dstImage)
plt.show()

# Visualizar histograma (dstImage)
hist = cv2.calcHist([dstImage],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```
