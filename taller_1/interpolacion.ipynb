```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.ones((256, 256), dtype=np.uint8) * 255

# Parámetros
grid_size = 8
thickness = 2
step = image.shape[0] // grid_size  # 32

# Dibujar líneas horizontales
for i in range(0, image.shape[0], step):
    image[i:i+thickness, :] = 0

# Dibujar líneas verticales
for j in range(0, image.shape[1], step):
    image[:, j:j+thickness] = 0

print(image.shape)
# Visualizar
plt.imshow(image, cmap="gray", vmin=0, vmax=255)
plt.show()
```

```python
# Submuestreo
image_resize =  cv2.resize(image, (128,128), interpolation=cv2.INTER_NEAREST)

print(image_resize.shape)
# Visualizar
plt.imshow(image_resize, cmap="gray", vmin=0, vmax=255)
plt.show()


# Submuestreo con 3 tecnicas de interpolacion
image_resize2 =  cv2.resize(image, (128,128), interpolation=cv2.INTER_LINEAR)

print(image_resize2.shape)
# Visualizar
plt.imshow(image_resize2, cmap="gray", vmin=0, vmax=255)
plt.show()

image_resize3 =  cv2.resize(image, (128,128), interpolation=cv2.INTER_AREA)

print(image_resize3.shape)
# Visualizar
plt.imshow(image_resize3, cmap="gray", vmin=0, vmax=255)
plt.show()

image_resize4 =  cv2.resize(image, (128,128), interpolation=cv2.INTER_CUBIC)

print(image_resize4.shape)
# Visualizar
plt.imshow(image_resize4, cmap="gray", vmin=0, vmax=255)
plt.show()
```

```python
# Imagen de trabajo
image_torax = cv2.imread("images/TORAX.png", cv2.IMREAD_GRAYSCALE)

# Visualizar imagen (TORAX)
plt.imshow(image_torax, cmap="gray")
plt.show()

image_resize =  cv2.resize(image_torax, (34,31), interpolation=cv2.INTER_NEAREST)

print(image_resize.shape)
# Visualizar
plt.imshow(image_resize, cmap="gray", vmin=0, vmax=255)
plt.show()

# Submuestreo con 3 tecnicas de interpolacion
image_resize2 =  cv2.resize(image_torax, (34,31), interpolation=cv2.INTER_LINEAR)

print(image_resize2.shape)
# Visualizar
plt.imshow(image_resize2, cmap="gray", vmin=0, vmax=255)
plt.show()

image_resize3 =  cv2.resize(image_torax, (34,31), interpolation=cv2.INTER_AREA)

print(image_resize3.shape)
# Visualizar
plt.imshow(image_resize3, cmap="gray", vmin=0, vmax=255)
plt.show()

image_resize4 =  cv2.resize(image_torax, (34,31), interpolation=cv2.INTER_CUBIC)

print(image_resize4.shape)
# Visualizar
plt.imshow(image_resize4, cmap="gray", vmin=0, vmax=255)
plt.show()
```

```python
# Ahora haga un remuestreo de la misma imagen
image_resize =  cv2.resize(image_torax, (134,124), interpolation=cv2.INTER_NEAREST)

print(image_resize.shape)
# Visualizar
plt.imshow(image_resize, cmap="gray", vmin=0, vmax=255)
plt.show()

# Duplico el valor con vecino mas cercano
image_resize_dupl =  cv2.resize(image_torax, (134*2,124*2), interpolation=cv2.INTER_NEAREST)

print(image_resize_dupl.shape)
# Visualizar
plt.imshow(image_resize_dupl, cmap="gray", vmin=0, vmax=255)
plt.show()

# Duplico el valor con cubica
image_resize_dupl_2 =  cv2.resize(image_torax, (134*2,124*2), interpolation=cv2.INTER_CUBIC)

print(image_resize_dupl_2.shape)
# Visualizar
plt.imshow(image_resize_dupl_2, cmap="gray", vmin=0, vmax=255)
plt.show()

# Esto lo hicimos para ver las difierencias, tal vez nuestro ojo humano no estaba notando la diferencia entre image_resize_dupl y image_resize_dupl_2
diff = cv2.absdiff(image_resize_dupl, image_resize_dupl_2)
plt.imshow(diff, cmap="hot")
plt.colorbar()
plt.title("Diferencia absoluta")
plt.show()
```
