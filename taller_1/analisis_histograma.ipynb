```python
import cv2
import matplotlib.pyplot as plt

image_muscle = cv2.imread("images/MUSCLE.png", cv2.IMREAD_GRAYSCALE)
image_clef = cv2.imread("images/CLEF.png", cv2.IMREAD_GRAYSCALE)
```

```python
# Visualizar imagen (MUSCLE)
plt.imshow(image_muscle, cmap="gray")
plt.show()

# Visualizar histograma (MUSCLE)
hist = cv2.calcHist([image_muscle],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
# Visualizar imagen (CLEF)
plt.imshow(image_clef, cmap="gray")
plt.show()

# Visualizar histograma (CLEF)
hist = cv2.calcHist([image_clef],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```
