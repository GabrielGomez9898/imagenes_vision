```python
import cv2
import matplotlib.pyplot as plt

image_rondelle = cv2.imread("images/RONDELLE.png", cv2.IMREAD_GRAYSCALE)
image_spot = cv2.imread("images/SPOT.png", cv2.IMREAD_GRAYSCALE)

# Visualizar imagen (RONDELLE)
plt.imshow(image_rondelle, cmap="gray")
plt.show()

# Visualizar imagen (SPOT)
plt.imshow(image_spot, cmap="gray")
plt.show()
```

```python
# Convertir a 16 bits las imagenes
image_rondelle = cv2.imread("images/RONDELLE.png", cv2.IMREAD_ANYDEPTH)
image_spot = cv2.imread("images/SPOT.png", cv2.IMREAD_ANYDEPTH)

# Visualizar histograma (image_rondelle)
hist = cv2.calcHist([image_rondelle],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

# Visualizar histograma (image_spot)
hist = cv2.calcHist([image_spot],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
# Adicionar rondelle.png a spot.png
image_addition = cv2.add(image_rondelle, image_spot)

# Visualizar imagen (image_addition)
plt.imshow(image_addition, cmap="gray")
plt.show()

# Visualizar histograma (image_spot)
hist = cv2.calcHist([image_addition],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
# Sustraccion rondelle.png a spot.png
image_subtract = cv2.subtract(image_rondelle, image_spot)

# Visualizar imagen (image_addition)
plt.imshow(image_subtract, cmap="gray")
plt.show()

# Visualizar histograma (image_subtract)
hist = cv2.calcHist([image_subtract],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```
