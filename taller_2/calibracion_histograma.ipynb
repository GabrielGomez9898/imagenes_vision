```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image_quito = cv2.imread("images/QUITO.png", cv2.IMREAD_GRAYSCALE)

# Visualizar imagen
plt.imshow(image_quito, cmap="gray",  vmin=0, vmax=255)
plt.show()

# Visualizar histograma
hist = cv2.calcHist([image_quito],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

print(image_quito.min())
print(image_quito.max())
```

Visualice el histograma de la imagen. Describa su forma. ¿Cuáles son los niveles de gris mínimo y máximo?.

Los niveles de gris es 35 y maximo 166

```python
# Expansion del contraste
from skimage import exposure

rescaled_img = exposure.rescale_intensity(image_quito, in_range=(35,166), out_range=(0,255))

# Visualizar imagen
plt.imshow(rescaled_img, cmap="gray", vmin=0, vmax=255)
plt.show()

# Visualizar histograma
rescaled_img = rescaled_img.astype(np.uint8)
hist = cv2.calcHist([rescaled_img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

Efectúe una expansión del contraste, visualice la imagen resultado y el histograma correspondiente. ¿En qué consiste la mejora de la imagen?

El objetivo de la expansion es mejorar el contraste, se ve mucho mejor las diferencias

```python
# Efectuar 3 calibraciones

# --- Calibrada 1 ---
rescaled_img = exposure.rescale_intensity(image_quito, in_range=(50,100), out_range=(0,255))

# Visualizar imagen
plt.imshow(rescaled_img, cmap="gray", vmin=0, vmax=255)
plt.title("Calibracion 1")
plt.show()

# Visualizar histograma
rescaled_img = rescaled_img.astype(np.uint8)
hist = cv2.calcHist([rescaled_img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

# --- Calibrada 2 ---
rescaled_img = exposure.rescale_intensity(image_quito, in_range=(40,120), out_range=(0,255))

# Visualizar imagen
plt.imshow(rescaled_img, cmap="gray", vmin=0, vmax=255)
plt.title("Calibracion 2")
plt.show()

# Visualizar histograma
rescaled_img = rescaled_img.astype(np.uint8)
hist = cv2.calcHist([rescaled_img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

# --- Calibrada 3 ---
rescaled_img = exposure.rescale_intensity(image_quito, in_range=(35,166), out_range=(0,255))

# Visualizar imagen
plt.imshow(rescaled_img, cmap="gray", vmin=0, vmax=255)
plt.title("Calibracion 3")
plt.show()

# Visualizar histograma
rescaled_img = rescaled_img.astype(np.uint8)
hist = cv2.calcHist([rescaled_img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

¿Cuál es la diferencia entre estas imágenes? ¿Cuál presenta el mejor contraste? ¿Cuál permite ver mejor los detalles? ¿Por qué la calibración entre 35 y 166 da el mismo resultado que la expansión del contraste efectuada en el punto 4? ¿Conclusión?

La diferencia es que la primera calibración al tener un rango mas pequeño, se truncan los demas pixeles, esto genera mas blancos y negros mucho mas resaltados.
A medida que se va ampliando el rango, se truncan menos pixeles y el rango de grises se aumenta. La mejor es la calibracion 1, me parece que es mucho mas evidente el contraste y se puede notar mas la diferencia. La calibracion 35 y 166 da el mismo resultado que el punto 4 porque son los valores minimos y maximos de la imagen.
