```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread("images/BOUGIES.png", cv2.IMREAD_GRAYSCALE)

# Visualizar imagen
plt.imshow(image, cmap="gray",  vmin=0, vmax=255)
plt.show()

# Visualizar histograma
hist = cv2.calcHist([image],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
# Mejorar el contraste
# Ecualizar el histograma
imagen_equ = cv2.equalizeHist(image)

# Visualizar imagen
plt.imshow(imagen_equ, cmap="gray",  vmin=0, vmax=255)
plt.show()
```

```python
# Umbralizacion
S=225
thres_value, umbral_image = cv2.threshold(imagen_equ, S, 255, cv2.THRESH_BINARY_INV)

# Visualizar imagen
plt.imshow(umbral_image, cmap='gray', vmin=0, vmax=255)
plt.title("Umbralización simple (S = 225)")
plt.show()

# Histograma
hist = cv2.calcHist([umbral_image],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
# Se hace un AND con la mascara recien hecha
result = cv2.bitwise_and(imagen_equ,umbral_image,0)

# Visualizar imagen
plt.imshow(result, cmap='gray', vmin=0, vmax=255)
plt.title("Resultado")
plt.show()

# Histograma
hist = cv2.calcHist([result],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```
