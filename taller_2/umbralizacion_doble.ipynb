```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image_rondelle = cv2.imread("images/RONDELLE.png", cv2.IMREAD_GRAYSCALE)

# Visualizar imagen
plt.imshow(image_rondelle, cmap="gray",  vmin=0, vmax=255)
plt.show()

# Visualizar histograma
hist = cv2.calcHist([image_rondelle],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

image_angio = cv2.imread("images/ANGIO.png", cv2.IMREAD_GRAYSCALE)

# Visualizar imagen
plt.imshow(image_angio, cmap="gray",  vmin=0, vmax=255)
plt.show()
```

```python
# Se desea aplicar una umbralización doble de tal manera que la imagen resultante debe contener tres clases, cuyos niveles de gris son 0, 128 y 255 respectivamente

# Imagen RONDELLE
S1 = 70
S2 = 160

result = np.zeros_like(image_rondelle, dtype=np.uint8)

# Clase alta
thres_value, mask_high = cv2.threshold(image_rondelle, S2, 255, cv2.THRESH_OTSU)
result[mask_high > 0] = 255

# Clase media
mask_mid = cv2.inRange(image_rondelle, S1, S2-1)
result[mask_mid > 0] = 128

# Visualizar imagen
plt.imshow(result, cmap='gray', vmin=0, vmax=255)
plt.title("thres_image_rondelle")
plt.axis('off')
plt.show()

# Histograma
hist = cv2.calcHist([result],[0],None,[256],[0,256])
plt.plot(hist)
plt.title("Histograma")
plt.show()
```

La umbralizacion doble, nos ayuda a identificar regiones, cuando es umbralizacion simple, nos deja solo blanco y negro y depronto perdemos informacion de algunas fronteras o regiones de la imagen. Conectividad, nos resalta con claridad las regiones, ya que los veicnos que son de colorees muy parecidos se clasifican en una misma categoria y ahi encontramos regiones.
